```python
import os
import shutil

# Tentukan parent folder dan folder tujuan
parent_folder = 'archive'  # Ganti dengan path ke parent folder Anda
destination_folder = 'Benchmark'
print(f"Folder tujuan: {destination_folder}")

# Buat folder 'Benchmark' jika belum ada
os.makedirs(destination_folder, exist_ok=True)

# Loop untuk menyalin semua gambar dari subfolder ke folder 'Benchmark'
for root, dirs, files in os.walk(parent_folder):
    for file in files:
        # Cek apakah file adalah gambar
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp', '.tiff', '.webp')):
            folder_name = os.path.basename(root)  # Nama folder asal
            new_file_name = f"{folder_name}_{file}"  # Nama file dengan format baru
            source_file = os.path.join(root, file)
            destination_file = os.path.join(destination_folder, new_file_name)
            
            # Salin file ke folder 'Benchmark'
            shutil.copy2(source_file, destination_file)
            print(f"Menyalin {file} sebagai {new_file_name} ke {destination_folder}")

print("Proses penyalinan selesai!")
```
